```python
import pandas as pd
import numpy as np
import seaborn as sns

#leer datos del dataset
datos = pd.read_csv('data/Celsius_Fahrenheit_dataset_1000.csv')
```

```python
# ver informacion del dataset
datos.info()
```

```python
# Mostramos los datos en un grafico de escala, visualizando que los datos dan como resultado una funcion lineal
sns.scatterplot(data=datos, x="Celsius", y="Fahrenheit", hue="Fahrenheit", palette="coolwarm")
```

```python
#preparar los datos para entrenar el modelo, Carateristicas (x) etiquetas (y)

x = datos["Celsius"]
y = datos["Fahrenheit"]

# agregar cada valor a un array para que el modelo pueda entenderlo
x_procesados = x.values.reshape(-1,1)
y_procesados = y.values.reshape(-1,1)

x_procesados
```

```python
# creamos el modelo
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
```

```python
#Entrenamos el modelo
modelo.fit(x_procesados, y_procesados)
```

```python
modelo.predict([[780]])
```

```python
#ver cual es el porciento de probabilidad es correcta la prediccion del modelo
modelo.score(x_procesados, y_procesados)
```
