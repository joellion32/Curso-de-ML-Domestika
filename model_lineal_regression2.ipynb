```python
#importar librerias
import pandas as pd
import seaborn as sms
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
#visualizar informacion sets de datos
data = pd.read_csv("data/SOCR-HeightWeight.csv")
data.info()
```

```python
#visualizar graficamente sets de datos
sms.regplot(data=data, x="Height", y="Weight", line_kws={"color": "red"})
```

```python
#preparar sets de datos adaptables al modelo
X = data["Height"]
Y = data["Weight"]

data_X = X.values.reshape(-1,1)
data_Y = Y.values.reshape(-1,1)
```

```python
#preparar modelo y entrenarlo
model = LinearRegression()
model.fit(data_X, data_Y)
```

```python
# visualizar promedio de exactiud de prediccion del modelo 
model.score(data_X, data_Y)
```

```python
#hacer prediccion
model.predict([[100]])
```
