```python
import pandas as pd
data = pd.read_csv("data/U4_01_housing.csv")
```

```python
#ver la cantidad de valores de una columna en especifico
data["ocean_proximity"].value_counts()
```

```python
#ver histogramas de datos es decir ver de manera grafica
data.hist(figsize=(15,8), bins=30, edgecolor="black")
```

```python
#crear grafica lineal con seaborn
import seaborn as sms
sms.scatterplot(x="latitude", 
                y="longitude", 
                data=data, 
                hue="median_house_value", 
                palette="coolwarm",
                s=data["population"] / 100)
```

```python
#eliminar filas con datos incompletos
data_na = data.dropna()
data_na.info()
```

```python
#como los modelos de algoritmos solo se entrenan con numeros la columna ocean_proximity puede afectar
#el entrenamiento de nuestro modelo, por eso lo que haremos sera convertir estos datos a datos numericos y utlizaremos el siguiente concepto
# por cada valor de la fila asinaremos un 1 y a los demas 0

#ejemplo

#Dummies / One Hot Coding 
# INLAND   NEAR BAY    NEAR OCEAN 
#  1         0            0 
#  0         1            0
#  0         0            1

#esto lo hacemos de la siguiente forma con pandas
dummies = pd.get_dummies(data["ocean_proximity"], dtype=int)
dummies
```

```python
#insertamos el nuevo arreglo de datos con a nuestro set de datos
data_na = data_na.join(dummies)
data_na
```

```python
#eliminamos la columna que tiene los datos que no son numericos
data_na = data_na.drop(["ocean_proximity"], axis=1)
data_na.head()
```

```python
# ver grafica de la correlacion de los datos
sms.set(rc={"figure.figsize": (15,8)})
sms.heatmap(data_na.corr(), annot=True, cmap="YlGnBu")
```

```python
# ver los datos correlacionados segun el precio de la casa
data_na.corr()["median_house_value"].sort_values(ascending=False)
```
